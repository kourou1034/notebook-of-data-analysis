```python
dx = 0.01
dy = 0.01
x = np.arange(-2.0, 2.0, dx)
y = np.arange(-2.0, 2.0, dy)
X,Y = np.meshgrid(x, y)
def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)
C = plt.contour(X,Y,f(X,Y),8,colors='black')
plt.contourf(X,Y,f(X,Y),8)
plt.clabel(C, inline=1, fontsize=10)
```

```python
gs = plt.GridSpec(3,2)
fig = plt.figure(figsize=(9,8))
x1 = np.arange(-3*np.pi, 3*np.pi, 0.01)
y1 = np.sin(2*x1)
s1 = fig.add_subplot(gs[0,:2])
s1.plot(x1, y1, linestyle='--', color='#FF1493', linewidth=3)
s1.spines['top'].set_color('none')
s1.spines['right'].set_color('none')
s1.spines['left'].set_position(('data',0))
s1.spines['bottom'].set_position(('data', 0))
s1.text(1.2,0.8,r'$y=sin(2x)$',fontsize=10, bbox={'facecolor':'white', 'alpha':0.2})
series1 = np.random.randint(2,10,4)
series2 = np.random.randint(2,10,4)
series3 = np.random.randint(2,10,4)
y2 = np.arange(4)
s2 = fig.add_subplot(gs[1,0])
s2.barh(y2, series1, color='w', hatch='*')
s2.barh(y2, series2, color='w', hatch='\\\\', left=series1)
s2.barh(y2, series3, color='w', hatch='//', left=(series1+series2))
labels = ['Python', 'Java', 'C++', 'C', 'Ruby']
values = [25, 35, 18, 15, 7]
colors = ['#FF1493', '#FFD700', '#00FF7F', '#DC143C', '#D8BFD8']
explode = [0.3, 0, 0, 0, 0]
s3 = fig.add_subplot(gs[1,1])
s3.pie(values, labels=labels, colors=colors, explode=explode, 
       autopct='%3.1f%%')
s3.axis('equal')
values2 = np.random.randint(20,30,10)
values3 = np.random.randint(30, 35, 10)
values4 = np.random.randint(15,22,10)
values5 = np.random.randint(10,16,10)
index = np.arange(10)
s4 = fig.add_subplot(gs[2,:2])
s4.plot(index, values2, linestyle='--',marker='o', color='#FF1493')
s4.plot(index, values3, linestyle='-',marker='>', color='#FFD700')
s4.plot(index, values4, linestyle='-.', color='#00FF7F')
s4.plot(index, values5, linestyle='-',marker='^', color='#DC143C')
s4.legend(['Python', 'Java', 'C++', 'C'], loc=0)
```
